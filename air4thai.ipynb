```
import folium # ติดตั้งได้ด้วย pip install folium
map_osm = folium.Map(location=[15.0000, 100.0000], zoom_start=5) # [ละติจูด,ลองติดจูด] และการซูมลำดับ 5
tooltip = 'Click me!'

data = air.get_data()

for i in data:
  if i['Lat']=='' and i['Lng']=='':
    continue
  #print([(i['Lat'].strip()), (i['Lng'].strip())])
  folium.Marker([float(i['Lat'].strip()), float(i['Lng'].strip())], popup=i['pm2.5']+" µg/m³", tooltip=tooltip).add_to(map_osm)
```

```
import xlsxwriter
import pandas as pd
```

```
# สร้าง DataFrame ที่มี 1 คอลัมน์ชื่อ 'สถานี'
dataframe = pd.DataFrame({'Station' : [i['title'] for i in air.get_data_provinces()['กรุงเทพฯ']],
                          'PM2.5' : [i['pm2.5'] for i in air.get_data_provinces()['กรุงเทพฯ']],
                          'PM10' : [i['pm10'] for i in air.get_data_provinces()['กรุงเทพฯ']],
                          'Lat' : [i['Lat'] for i in air.get_data_provinces()['กรุงเทพฯ']],
                          'Lng' : [i['Lng'] for i in air.get_data_provinces()['กรุงเทพฯ']],
                          'time' : [i['time'] for i in air.get_data_provinces()['กรุงเทพฯ']]})
```

```
# สร้าง Pandas Excel Writer เพื่อใช้เขียนไฟล์ Excel โดยใช้ Engine เป็น xlsxwriter
# โดยตั้งชื่อไฟล์ว่า 'simple_data.xlsx'
writer = pd.ExcelWriter('air4thai.xlsx', engine='xlsxwriter')
 
```

```
# นำข้อมูลที่สร้างไว้ในตัวแปร dataframe เขียนลงไฟล์
dataframe.to_excel(writer, sheet_name='หน้าที่1')
```

```
# จบการทำงาน Pandas Excel writer และเซฟข้อมูลออกมาเป็นไฟล์ Excel
writer.save()
```
